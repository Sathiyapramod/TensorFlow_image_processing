```python
import gradio as gr
from transformers import pipeline

classifiers = pipeline("text-classification", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")

def predict_sentiment(text:str):
  return classifiers(text)[0]

my_interface = gr.Interface(
    fn=predict_sentiment,
    title="Sentiment Analysis",
    inputs="textbox",
    outputs="textbox"
)

my_interface.launch(share=True)
```
